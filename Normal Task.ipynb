### Exploratory Data Analysis (EDA)

Descriptive statistics: 
mean, mode, median, count: The mean and median purchase amounts provide insights into the average spending behaviour of customers.

#### Exploring the value column

Value column represents the Purchase amount of the products

```python
# Calculate mean purchase amount
mean = df['VALUE'].mean()
print("The Mean purchase amount is:",mean)
```

```python
# Calculate mediam purchase amount
median = df['VALUE'].median()
print("The Median purchase amount is:",median)
```

```python
# Calculate standard deviation of purchase amount
std = df['VALUE'].std()
print("The Standard deviation of purchase amount is:",std)
```

#### Exploring the Quantity column

```python
# Calculate mean Quantity
mean = df['QTY'].mean()
print("The Mean of Quantity is:",mean)
```

```python
# Calculate median Quantity
median = df['QTY'].median()
print("The Median of Quantity is:",median)
```

```python
# Calculate standard deviation of Quantity
std = df['QTY'].std()
print("The Standard deviation of Quantity is:",std)
```

By this Visualizing the Graph we can conclude that M3 (i.e Month 3) Has more transactions than the remaining.

##### Plotting Each Store With Quantity

```python
max_quantity_bystore = df.groupby('STORECODE')['QTY'].sum().sort_values(ascending = False)
max_quantity_bystore
```

```python
max_quantity_bystore = df.groupby('STORECODE').sum().sort_values("QTY",ascending = False)
```

```python
# Plotting each store with value
plt.figure(figsize = (12, 6))
plt.title('Top Stores with highest quantities sold')
sns.barplot(x = 'QTY', y = max_quantity_bystore.index , data = max_quantity_bystore)
```

```python
for i in df.STORECODE.unique():
    sample_data=df.loc[:,['STORECODE','GRP','QTY']]
    top_store=sample_data.loc[sample_data.STORECODE == i].groupby("GRP").sum()
    top_store=top_store.sort_values(by='QTY', ascending=False)
    top_store=top_store.iloc[0:5]
    plt.figure(figsize=(12,6))
    num_categories = 5
    palette = sns.color_palette('hsv', num_categories)
    sns.barplot(x='QTY',y=top_store.index,data=top_store)
    plt.title("Top 5 Categories Sold from "+ i+ " Store")
    plt.xlabel("Number of Quantities")
    plt.ylabel("Categories")
    plt.grid(axis='x',color='black')
```

```python
for i in df.MONTH.unique():
    sample_data=df.loc[:,['MONTH','GRP','QTY']]
    month_fig=sample_data.loc[sample_data.MONTH == i].groupby("GRP").sum()
    month_fig=month_fig.sort_values(by='QTY', ascending=False)
    month_fig=month_fig.iloc[0:5]
    plt.figure(figsize=(12,6))
    num_categories = 5
    palette = sns.color_palette('hsv', num_categories)
    sns.barplot(x='QTY',y=month_fig.index,data=month_fig)
    plt.title("Top 5 Categories Sold by Month "+ i)
    plt.xlabel("Number of Quantities")
    plt.ylabel("Categories")
    plt.grid(axis='x',color='black')
```
